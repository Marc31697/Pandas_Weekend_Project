```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('AB_NYC_2019.csv')
air_bnb.head()
```

```python
# How many neighborhood groups are available and which shows up the most?

air_bnb.groupby('neighbourhood_group')['neighbourhood_group'].count().sort_values(ascending=False).to_frame()

# 5 Groups, Manhattan shows up the most
```

```python
# Are private rooms the most popular in manhattan?

display(air_bnb[air_bnb['neighbourhood_group'] == 'Manhattan'].groupby(['neighbourhood_group','room_type'])['room_type'].count().to_frame()),

display(air_bnb[air_bnb['neighbourhood_group'] == 'Manhattan'].groupby(['neighbourhood_group','room_type'])['number_of_reviews'].sum().to_frame())

# There are more Entire Home/apt room types and more reviews for that type than there are for Private Rooms
# Therefore Private Rooms are not the most popular in manhattan
```

```python
# Which hosts are the busiest based on their reviews?

display(air_bnb.groupby(['host_id', 'host_name'])['reviews_per_month'].sum().sort_values(ascending=False).head(10).to_frame())

# Sonder appears to be the busiest as they have the most reviews per month
```

```python
#Which neighorhood group has the highest average price?

display(air_bnb.groupby(['neighbourhood_group'])['price'].mean().sort_values(ascending=False).to_frame())

# Manhattan has the highest average price
```

```python
# Which neighbor hood group has the highest total price?

air_bnb[ ['neighbourhood_group', 'price'] ].sort_values('price', ascending = False).head(3)

# Appears to be a tie between Queens, Brooklynn, and Manhattan
```

```python
#Which top 5 hosts have the highest total price?

air_bnb[ ['host_id', 'host_name', 'price']].sort_values('price', ascending = False).head(5)

# Katherine -> Erin -> Jelena -> Matt -> Amy
```

```python
# Who currently has no (zero) availability with a review count of 100 or more?

air_bnb.loc[(air_bnb['availability_365'] == 0) & (air_bnb.groupby('host_id')['number_of_reviews'].sum() >= 100)][ ['host_id', 'host_name'] ]

# Listed Below
```

```python
# What host has the highest total of prices and where are they located?

air_bnb.groupby(['host_id', 'host_name', 'neighbourhood_group'])['price'].sum().sort_values(ascending=False).to_frame()

# Sonder has the highest total of prices and they are located in Manhattan
```

```python
# When did Danielle from Queens last receive a review?

air_bnb[(air_bnb['host_name'] == 'Danielle') & (air_bnb['neighbourhood_group'] == 'Queens')][ ['host_id','host_name','neighbourhood_group','last_review'] ]

# Depends on which Danielle
```
